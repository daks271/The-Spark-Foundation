### There is no null values in our dataset

```python
# To check the duplicate records
duplicate_rows = df[df.duplicated()]
print('Number of duplicate rows: ', duplicate_rows.shape)
```

```python
# Drop the duplicate records
df1 = df.drop_duplicates().reset_index(drop=True)
print("Shape of the data:",df1.shape)
df1.head() 
```

### Exploratory Data Analysis

```python
df1.nunique()
```

### From above "Country" has only one unique value and "Postal Code" is not necessary for our analysis so we drop these columns

```python
# Drop these column
df1 = df1.drop(columns={'Country','Postal Code'},axis=1)
df1.head()
```

```python
# Group by
category_sales_profit = df1.groupby('Category').sum()
display(category_sales_profit)
```

```python
# Sales,Quantity,Discount and profit of categories
plt.style.use('seaborn-darkgrid')
#colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
category_sales_profit.plot.pie(colors = colors,shadow = True, startangle=90, subplots=True,figsize=(50,20),autopct='%1.1f%%')
plt.rcParams['font.size'] =25.0
plt.show()
```

### From The Above Bar Chart :

#### For total Sales % By Technology is 36.4% & Quantity is 18.3% , Discount is 15.7% and Profit is 50.8%
#### For total Sales % By Office Supplies is 31.3% & Quantity is 60.4% , Discount is 60.7% and Profit is 42.7%
#### For total Sales % By Furniture is 32.3% & Quantity is 21.2% , Discount is 23.6% and Profit is 6.4%

```python
Subcategory_sales_profit = df1.groupby('Sub-Category').sum()
display(Subcategory_sales_profit)
```

```python
#Sales and Profits of Categories and Sub-Categories
fig=plt.figure(figsize=(15,8))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
df1.groupby("Category")[["Sales","Profit"]].sum().sort_values(by="Sales",ascending=False).plot(kind="bar",ax=ax1)
ax1.set_title("Sales and Profits for each Category",size=18)
df1.groupby("Sub-Category")[["Sales","Profit"]].sum().sort_values(by ="Sales",ascending=False).plot(kind ="bar",ax=ax2)
ax2.set_title("Sales and Profits for each Sub-Category",size=18)
plt.show()
```

### From plot 1 :

In Category the maximum profit is for Technology with sales around 836K and the minimum Profit for category Furniture with sales around 741K

### From plot 2 :

In Subcategory the maximum profit is for Copiers with sales nearly 150K and the profit is negative for sub category Tables with sales nearly 206K

```python
# group by quantity , sales and profit
Quantity_sales_profit = df1.groupby('Quantity').sum() 
display(Quantity_sales_profit)
```

```python
Segment_sales_profit=df1.groupby('Segment').sum()
display(Segment_sales_profit)
```

```python
#Sales and Profits of each Quantity and each Segment
fig=plt.figure(figsize=(15,8))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(222)
df1.groupby("Quantity")[["Sales", "Profit"]].sum().sort_values(by = "Sales", ascending = False).plot(kind = "bar",ax=ax1)
ax1.set_title("Sales and Profits for each Quantity",size=18)
df1.groupby("Segment")[["Sales", "Profit"]].sum().sort_values(by = "Sales", ascending = False).plot(kind = "bar",ax=ax2)
ax2.set_title("Sales and Profits for each Segment",size=18)
plt.show()
```

### From plot 1 :

The sales is maximum for lesser quantities and so is the profit.

### From plot 2 :

The maximum profit in *Segment the consumer having maximum sales and the profit is minimum for segment Home office** with minimum sales amongst the segment.

```python
#Sales and Profits of each State
fig=plt.figure(figsize=(50,20))
ax1=fig.add_subplot(121)
df1.groupby("State")[["Sales","Profit"]].sum().sort_values(by = "Sales", ascending = False).plot(kind = "bar",ax=ax1)
ax1.set_title("Sales and Profits for each State", size=28)
plt.show()
```

### From above :

for states including Texas,Pennsylvania, Florida, Illinois, Ohio, North Carolina and few other states have high sales with negative profits.

```python
#statewise discount
df1.groupby("State").sum()['Discount'].sort_values(ascending = False)
```

#### From above :

We can clearly observe that discount has a negative relation with Profit i.e. as the discount is increasing the Profit generated is decreased.

### Data with respect to negative profits

```python
Negative_profits=df1.loc[df1.Profit <= 0].sort_values(by = "Profit",ascending = True).reset_index(drop=True)
Negative_profits.head(10)
```

```python
#Negative Profit by Category
by_cat=Negative_profits.groupby('Category').sum().reset_index()
by_cat
```

```python
plt.subplots(figsize=(10,2))
sns.barplot(y=by_cat.Category,x=by_cat.Profit)
plt.title("Negative Profit by Category",size=20)
plt.show()
```

```python
#Negative Profits for each Category
fig=plt.figure(figsize=(50,8))
ax1=fig.add_subplot(121)
Negative_profits.groupby("Category")[["Sales","Profit","Quantity"]].sum().sort_values(by = "Sales", ascending = False).plot(kind = "bar",ax=ax1)
ax1.set_title("Negative Profits for each Category",size=30)
plt.show()
```

### From the above two bar plots, we observe that sales which are maximum nearly 276K is causing the most negative profits nearly -60K.

```python
# Negative Profit by City
by_city=Negative_profits.groupby('City').sum().sort_values(by='Profit').reset_index()
by_city.head()
```

```python
#Top 5 Cities with most negative profit
plt.subplots(figsize=(10,2))
sns.barplot(y=by_city.head().City,x=by_city.head().Profit)
plt.title("Top 5 Cities with most negative profit",size=20)
plt.show()
```
